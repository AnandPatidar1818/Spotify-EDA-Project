* **Peak & Volatility**

### **12. Combined Ranking of Top Artists**

```python
artist_score = df.groupby('artist_name').agg({
    'track_popularity':'mean',
    'artist_popularity':'mean',
    'artist_followers':'mean'
})


artist_score['combined_score'] = (
    artist_score['track_popularity'] +
    artist_score['artist_popularity'] +
    artist_score['artist_followers']
)

artist_score = artist_score / 100

artist_score.sort_values('combined_score', ascending=False).head(5)
```

### Combined Artist Score Analysis

* A **combined score** was created using:

  * Track popularity
  * Artist popularity
  * Artist followers

* **Top-scoring artists are very few** â†’ long-tail concentration.

* **Followers contribute most** to the combined score because values are in millions.

* **Artist popularity score is normalized (0â€“1)** while followers are **not scaled**, causing imbalance.

* Ranking shows **audience reach dominance**, not recent engagement dominance.

* Combined score helps identify **platform giants**, but may hide **high-engagement low-follower artists**.

### **13. Conclusion and Insights**


### Final Conclusion

* The dataset contained **millions of follower values vs normalized popularity scores**, requiring careful scaling to avoid bias.

* Spotifyâ€™s music ecosystem shows **unequal audience distribution**, where a small group dominates total followers and combined scores.

* **Track popularity is influenced more by artist popularity than artist followers**, proving that recent streaming engagement matters more than long-term fan count for trending tracks.

* Most tracks follow a **standard duration of 3â€“5 minutes**, forming a normal distribution, but **duration itself is not a strong predictor of popularity**.

* **Explicit tracks do not dominate popularity**, but some high-popularity outliers show that explicit content can still perform well when supported by strong artist engagement or playlisting.

* Music releases grew **slowly before 2000**, then **exponentially increased after 2010**, reflecting the rise of streaming platforms and independent artists.

* The heatmap confirmed that:

  * **Artist popularity correlates moderately with followers (~0.64)**.
  * **Album size has almost no effect on track popularity (-0.054)**.
  * **Track popularity has weak correlation with followers (0.23)**.

* The **Top 5 ranking varied by metric**, meaning:
  * Some artists are **productive but not viral**.

  * This supports your objective that **Spotify success is multi-factor driven, not single-metric dependent**.

---

### Core Insights

* **Spotify popularity is momentum-based**, not length-based.

* **Followers represent loyalty, popularity score represents recency and engagement** â€” both must be interpreted separately.

* **Promotion format matters**: singles benefit most from algorithmic recommendations and playlists.

* **Music production is democratized** â€” more artists release frequently, especially in recent years.

* **Fame inequality exists**, but engagement can rise even without huge followers.

* **Data scaling changes conclusions** â€” combined scores must be log or normalized for fair ranking.
